## Now let's stop with descriptive statistics and let's move on to inferential statistics.

## Let's predict how many police deaths might occur in 2022. I will not use the entire dataset for prediction in this case. Let's start with the samples. Since we are predicting 2022, we will take samples from the years 2000 to 2021.

## Since we are dealing with confidence interval i will make two function one for t - test and other of z - test. We might need both of them.

```python
# 95% confidence level is considered normally but you can change it to 90% or 99% as you needed.
# Make sure to pass the dataframe array of a required table like i will do it ahead

def confidence_interval_t_test(df_array, confidence_level = 0.95):
 
    description = df_array.describe() # Here I will use describe method to get count, mean and standard deviation of the data.
    
    n = round(description[0]) # Count of the array
    
    df = n - 1 # Degree of freedom
    
    x_bar = round(description[1]) # Mean of the data
    
    s = round(description[2]) # Standard Deviation of the data
    
    alpha = 1 - confidence_level # Alpha value or Significance level
    
    standard_error = s / math.sqrt(n) # Standar error formula
    
    critical_t_value = stats.t.ppf(1 - alpha/2, df) # Critical t-value using stats from scipy
    
    margin_of_error = critical_t_value * standard_error # Margin of Error formula
    
    lower_bound = x_bar - margin_of_error # Calculating lower bound
    
    upper_bound = x_bar + margin_of_error # Calculating Upper bound
    
    confidence_interval = [round(lower_bound), round(upper_bound)] # Returning in the list form
    
    return confidence_interval
```

```python
# 95% confidence level is considered normally but you can change it to 90% or 99% as you needed.
# Make sure to pass the dataframe array of a required table like i will do it ahead
# For z test statistic we don't need degree of freedom

def confidence_interval_z_test(df_array, confidence_level = 0.95):
 
    description = df_array.describe() # Here I will use describe method to get count, mean and standard deviation of the data.
    
    n = round(description[0]) # Count of the array
    
    x_bar = round(description[1]) # Mean of the data
    
    s = round(description[2]) # Standard Deviation of the data
    
    alpha = 1 - confidence_level # Alpha value or Significance level
    
    standard_error = s / math.sqrt(n) # Standar error formula
    
    critical_z_value = stats.norm.ppf(1 - alpha/2) # Critical z-value using stats from scipy
    
    margin_of_error = critical_t_value * standard_error # Margin of Error formula
    
    lower_bound = x_bar - margin_of_error # Calculating lower bound
    
    upper_bound = x_bar + margin_of_error # Calculating Upper bound
    
    confidence_interval = [round(lower_bound), round(upper_bound)] # Returning in the list form
    
    return confidence_interval
```

## I will predict deaths for human unit, K9 unit and for both but before that i need to make a dataframe consist of all this information.

```python
# Human Unit count
humans_df = police_deaths_df[police_deaths_df["K9_Unit"]==0].reset_index(drop=True)
humans_df = humans_df.drop("K9_Unit",axis=1)

# Making Count df
l = ['Name','Cause_of_Death','Date','Month','Day','Department','State']
count_df = humans_df.groupby(by="Year").count().drop(l,axis=1)
count_df = count_df.rename(columns={'Rank':'Count_Human_Unit'})

# Transfering to all unit death counts df
all_unit_count = count_df.tail(23)
all_unit_count = all_unit_count.drop(index=2022,axis=0)

# K9 Unit count
k9_df = police_deaths_df[police_deaths_df["K9_Unit"]==1].reset_index(drop=True)
k9_df = k9_df.drop("K9_Unit",axis=1)

# Making Count df
l = ['Name','Cause_of_Death','Date','Month','Day','Department','State']
k9_count_df = k9_df.groupby(by="Year").count().drop(l,axis=1)
k9_count_df = k9_count_df.rename(columns={'Rank':'Count_K9_Unit'})

# Filtering years
fil_k9_count_df = k9_count_df.tail(23)
fil_k9_count_df = fil_k9_count_df.drop(index=2022,axis=0)

# Transfering to all unit death counts df
all_unit_count["Count_K9_Unit"] = fil_k9_count_df["Count_K9_Unit"]

# Making a Total column
all_unit_count["Combined"] = all_unit_count["Count_Human_Unit"] + all_unit_count["Count_K9_Unit"]

all_unit_count
```

## Now i will pass these three arrays 1 by 1 and make a new dataframe called predicted deaths for 2022.

## We will carry these tests with the help of the t-test statistic because we don't have enough samples. For example, here we have n = 22 since we are taking data from 2020 to 2021, including both years.

```python
pred_list = []

for items in all_unit_count.columns:
    pred = confidence_interval_t_test(all_unit_count[items])
    pred_list.append(pred)
    
predicted_deaths_2022 = pd.DataFrame(index = ["Human_Unit","K9_Unit","Combined"])

predicted_deaths_2022["Deaths_Estimation_at_95%_Confidence_Level"] = pred_list

predicted_deaths_2022
```

## With this test, we can estimate that in 2022 at 95% confidence level :-
## Human police unit deaths will be approximately, between 156 and 252, 
## K9 unit deaths will be approximately between 12 and 20, and 
## Combined police unit deaths will be aprroximately between 171 and 269.
